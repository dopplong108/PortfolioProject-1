```python
#  Count and drop duplicates

dup_rows = df.duplicated(keep = 'first').sum()
print(dup_rows)
```

```python
df_cleaned = df.drop_duplicates(keep = 'first')
df_cleaned.head()
```

```python
# sorting data by cust_state and formating the trans_timestamp to date only
df_cleaned.sort_values(by = 'cust_state', ascending = True)
```

```python
df_cleaned.head()
```

```python
# remove any rows with values greater than 1000 in prod_price and sve as new data
df_num_fixed = df_cleaned[~(df_cleaned['prod_price'] >= 1000)]
```

```python
# Correct the values in total_sales by using prod_price and trans_quantity
df_cleaned['total_sales'] =  df_cleaned['prod_price'] * df_cleaned['trans_quantity']
```

```python
# Finding missing values
for a in df_cleaned:
    pct_missing = np.mean(df_cleaned[a].isnull())
    print('{} - {}%'.format(a, pct_missing*100))
# Deleting missing value
df_checked = df_cleaned.dropna()
df_checked.isnull().sum()
```
